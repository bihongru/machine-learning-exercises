### Step 1: Import Libraries

```python
import cv2
import matplotlib.pyplot as plt
```

### Step 2: Read Image

```python
img = cv2.imread('./data/Screenshot.png')
```

### Step 3: Prep Grayscale & Blur

```python
g = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
g = cv2.medianBlur(g, 5)
```

```python
plt.imshow(g, cmap=plt.cm.binary)
```

### Step 4: Edges

```python
e = cv2.adaptiveThreshold(g, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 9)
```

```python
plt.imshow(e, cmap=plt.cm.binary)
```

### Step 5: Smooth Color

```python
c = cv2.bilateralFilter(img, 9, 250, 250)
```

### Step 6: Combine

```python
cartoon = cv2.bitwise_and(c, c, mask=e)
```

### Step 7: Show Result

```python
fig, ax = plt.subplots(2, 1, figsize=(12, 6))
ax[0].imshow(img)
ax[0].axis('off')
ax[1].imshow(cartoon)
ax[1].axis('off')
```
