### Step 1: Import Libraries

```python
import cv2
import pytesseract
import matplotlib.pyplot as plt
```

### Step 2: Loading Image

```python
image_path = './data/Original-Image.png'
```

```python
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image_rgb)
```

### Step 3: Converting Image to Grayscale

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray, cmap=plt.cm.Grays_r)
```

### Step 4: Set Tesseract Path

```python
pytesseract.pytesseract.tesseract_cmd = 'E:/software/Tesseract-OCR/tesseract.exe'
```

### Step 5: Extracting Text from Image

```python
extracted_text = pytesseract.image_to_string(gray)
```

```python
print(extracted_text)
```

### Step 6: Drawing Bounding Boxes Around Detected Text

```python
data = pytesseract.image_to_data(gray, output_type=pytesseract.Output.DICT)
```

```python
n_boxes = len(data['level'])
```

```python
for i in range(n_boxes):
    (x, y, w, h) = (data['left'][i], data['top'][i], data['width'][i], data['height'][i])
    cv2.rectangle(image_rgb, (x, y), (x+w, y+h), (0, 255, 0), 1)
```

```python
plt.imshow(image_rgb)
```
