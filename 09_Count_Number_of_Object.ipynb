### Step 1: Import Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Step 2: Read Image

```python
image = cv2.imread('./data/bitss.jpg')
```

```python
plt.imshow(image[:,:,::-1])
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray, cmap=plt.cm.Grays_r)
```

### Step 3: Blur Image

```python
blur = cv2.GaussianBlur(gray, (11, 11), 0)
```

```python
plt.imshow(blur, cmap=plt.cm.Grays_r)
```

### Step 4: Detect Edges

```python
canny = cv2.Canny(blur, 30, 150, 3)
```

```python
plt.imshow(canny, cmap=plt.cm.Grays_r)
```

### Step 5: Connect the Edges

```python
dilated = cv2.dilate(canny, (1, 1), iterations=0)
```

```python
plt.imshow(dilated, cmap=plt.cm.Grays_r)
```

### Step 6: Calculate the Contour in the Image

```python
(cnt, hierarchy) = cv2.findContours(dilated.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
```

```python
rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
_ = cv2.drawContours(rgb, cnt, -1, (0, 255, 0), 2)
```

```python
plt.imshow(rgb)
```
