### Step 2: Tokenize, Stemize, Lemmatize Data

```python
df = pd.read_csv('./data/all_kindle_review.csv')
```

```python
text = df['reviewText'].values
```

```python
text = text.tolist()
```

```python
text = "i love my kindle"
```

```python
sent_tokenizer = PunktSentenceTokenizer(text[0])
```

```python
sents = sent_tokenizer.tokenize(text[0])
```

```python
sents
```

```python
nltk_tokens = nltk.word_tokenize(text[0])
```

```python
print(word_tokenize(sents[0]))
print(sent_tokenize(sents[0]))
```

```python
porter_stemmer = PorterStemmer()
```

```python
wordnet_lemmatizer = WordNetLemmatizer()
```

```python
for w in nltk_tokens:
    print("Actual: % s Stem: % s" % (w, porter_stemmer.stem(w)))
    print("Actual: % s Lemma: % s" % (w, wordnet_lemmatizer.lemmatize(w)))
    print('-----------------')
```

```python
print(nltk.pos_tag(nltk_tokens))
```
