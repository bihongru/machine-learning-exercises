### Step 1: Import Libraries

```python
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
```

### Step 2: Preparing Dataset

```python
digits = datasets.load_digits()
```

```python
dir(digits)
```

### Step 3: Function to Print a Set of Image

```python
digits.images[0]
```

```python
def plot_multi(i):
    nplots = 16
    fig = plt.figure(figsize=(10, 10))
    for j in range(nplots):
        plt.subplot(4, 4, j+1)
        plt.imshow(digits.images[i+j], cmap=plt.cm.binary)
        plt.title(digits.target[i+j])
        plt.axis('off')
```

```python
plot_multi(0)
```

### Step 4: Training Neural Network with the Dataset

```python
y = digits.target
x = digits.images.reshape(len(digits.images), -1)
```

```python
x.shape
```

### Step 5: Data Split for Training and Testing

```python
x_train = x[:1000]
y_train = y[:1000]
```

```python
x_test = x[1000:]
y_test = y[1000:]
```

### Step 6: Usage of Multi-Layer Perceptron Classifier

```python
mlp = MLPClassifier(
    hidden_layer_sizes=(15,),
    activation='logistic',
    alpha=1e-4,
    solver='sgd',
    tol=1e-4,
    random_state=1,
    learning_rate_init=.1,
    verbose=True
)
```

```python
mlp.fit(x_train, y_train)
```

```python
fig, axes = plt.subplots(1, 1)
axes.plot(mlp.loss_curve_, 'o-')
axes.set_xlabel("number of iteration")
axes.set_ylabel("loss")
```

### Step 7: Model Evaluation

```python
predictions = mlp.predict(x_test)
```

```python
accuracy_score(y_test, predictions)
```
